```python
# WHOLE CODE IN A SINGLE CELL TO SAVE TIME ON EXECUTION
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time

trends = TrendReq()
search_term = input("What search term are you looking for?   ")




trends.build_payload(kw_list=[search_term])

data = trends.interest_by_region()
data_all = data
# data = trends    # CHECK FOR OTHER DEFAULT .interest_by_XXXXXXX as there are multiple options to explore
# print(type(data))

data = data.sort_values(by=search_term, ascending=False)
data = data.head(10)
print(data)
#print(data.columns)
#print(data.index)



# find a specific country's position in the current ranking
print("")
country = input("Position of which country do you want to check?   ")
data_all = data_all.sort_values(by=search_term, ascending=False)
data_all["Rank"] =  range(1, len(data_all)+1) 
rank = data_all.loc[country, "Rank"]
print(f"{country} => {rank}")
print("")
# Set the display options to show all the rows and columns
pd.set_option('display.max_rows', None)
print(data_all)
print("")



data.reset_index().plot(x="geoName",
                        y=search_term,
                        figsize=(15,12),
                        kind="bar")
plt.style.use("fivethirtyeight")
plt.show()



data = TrendReq(hl="en-US", tz=360)
data.build_payload(kw_list=[search_term])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15,12))
data[search_term].plot()
plt.style.use("fivethirtyeight")
plt.title(f"Total Google Searches for {search_term}",
          fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Total Count")
plt.show()
```
